### Wikidex dataset visualization

```python
stats = wikidex[["National Dex #", "HP", "Attack", "Defense",
                 "Special Attack", "Special Defense", "Speed"]].copy()
pd.set_option('display.max_rows', None)
wikidex
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(
    rows=6, cols=1,
    subplot_titles=("HP", "Attack", "Defense", "Special Attack", "Special Defense", "Speed"))

fig.add_trace(go.Histogram(x=stats["National Dex #"], y=stats["HP"], name="HP", xbins=dict(start=1,end=1021,size=1), ybins=dict(start=1,end=255,size=1)),
              row=1, col=1)

fig.add_trace(go.Histogram(x=stats["National Dex #"], y=stats["Attack"], xbins=dict(start=1,end=1021,size=1)),
              row=2, col=1)

fig.add_trace(go.Histogram(x=stats["National Dex #"], y=stats["Defense"], xbins=dict(start=1,end=1021,size=1)),
              row=3, col=1)

fig.add_trace(go.Histogram(x=stats["National Dex #"], y=stats["Special Attack"], xbins=dict(start=1,end=1021,size=1)),
              row=4, col=1)

fig.add_trace(go.Histogram(x=stats["National Dex #"], y=stats["Special Defense"], xbins=dict(start=1,end=1021,size=1)),
              row=5, col=1)

fig.add_trace(go.Histogram(x=stats["National Dex #"], y=stats["Speed"], xbins=dict(start=1,end=1021,size=1)),
              row=6, col=1)

fig.update_layout(height=700, width=900,
                  title_text="Multiple Subplots with Titles")

fig.show()
```
