## Wikidex dataset

```python
wikidex = pd.read_csv("data/wikidexDF_clean.csv")
```

```python
wikidex.shape
```

```python
wikidex.info()
```

```python
wikidex.head()
```

```python
wikidex.tail()
```

```python
wikidex = wikidex.astype({"HP": "float64", "Speed": "float64"})
wikidex["Attack"] = wikidex["Attack"].replace(",",".").replace(" ","").replace("140 35","140").astype(float)
wikidex["Defense"] = wikidex["Defense"].replace(",",".").replace(" ","").replace("140 35","140").astype(float)
wikidex["Special Attack"] = wikidex["Special Attack"].replace(",",".").replace(" ","").replace("140 35","140").astype(float)
wikidex["Special Defense"] = wikidex["Special Defense"].replace(",",".").replace(" ","").replace("140 35","140").astype(float)
wikidex["Mean"] = wikidex["Mean"].str.replace(",",".").replace(" ","").replace("140 35","140").astype(float)
wikidex["SD"] = wikidex["SD"].str.replace(",",".").replace(" ","").replace("140 35","140").astype(float)
```
